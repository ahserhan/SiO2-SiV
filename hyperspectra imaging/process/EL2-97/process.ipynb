```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm
from lmfit.models import QuadraticModel, LorentzianModel, LinearModel
from lmfit import Model, Minimizer, Parameters, report_fit
import math

def findarg(data, value):
    arg = 0
    for wl in data:
        if wl<value:
            arg = arg + 1
        else:
            return arg
        
    return arg


def integrate (wavelength, intensity, minWavelength, maxWavelength):
    sum = 0
    Index = list(range(findarg(wavelength, minWavelength),findarg(wavelength, maxWavelength)))
    for i in Index:
        sum += intensity[i]*(wavelength[i+1]-wavelength[i])
    return sum

data = pd.read_pickle('asher_20230603_hyperspec_EL2_97.pkl')
ramanData = pd.read_csv('secondRaman.csv')

ramanI = np.array(ramanData['I'])
im = data['im']
wavel = data['wavelength']
map = np.sum(im, axis = 2)
# pixel numbers x and y
px = len(im)
py = len(im[0])
ratiointensity = np.zeros((px,py))
confocalintensity = np.zeros((px,py))
nvintensity = np.zeros((px,py))
sivintensity = np.zeros((px,py))
r2intensity = np.zeros((px,py))
```

# Bin all

```python
#total intensity plot
plt.figure(figsize=(12,9))
plt.grid()
plt.imshow(map, cmap='viridis', vmax = 3e6)
plt.colorbar()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

# Bin 580 - 850

```python
#bin selection min and max
binMin = 580
binMax = 850
# binMin = 760
# binMax = 800
argmin = findarg(wavel, binMin)
argmax = findarg(wavel, binMax)
#integrate over the bin
for i in range(px):
    for j in range(py):
        confocalintensity[i][j] = im[i][j][argmin:argmax].sum()
#bin intensity plot
plt.figure(figsize=(12,9))
plt.imshow(confocalintensity, cmap='viridis', vmax = 2.5e6)
plt.colorbar()
plt.show()
```

# Bin NV-

```python
def getNvIntensity(data, regionx, regiony):
    currentPL = data['im'][regionx][regiony]
    model2 = QuadraticModel(prefix='bkg_') + LorentzianModel(prefix='nvm_')
    model = model2
    params = model.make_params()

    params['nvm_center'].set(value=639, min = 638, max = 640, vary=True)
    params['nvm_amplitude'].set(value=1000, min = 0, vary=True)
    params['nvm_sigma'].set(value=2, max = 3, vary=True)

    params['bkg_a'].set(value = 0, vary=True)
    params['bkg_b'].set(value = 0, min = 0, vary=True)
    params['bkg_c'].set(value = 0, vary=True)

    Index = list(range(910,1110))
    weightList = 1/np.sqrt(currentPL[Index])
    result = model.fit(currentPL[Index], params, x=wavel[Index], weights=weightList)
    comps = result.eval_components(x=wavel[Index])

    nvmCenter = result.params['nvm_center']
    nvmHWHM = result.params['nvm_sigma']
    print(regionx,regiony)
    return integrate(wavel[Index],comps['nvm_'],nvmCenter - nvmHWHM*2, nvmCenter + nvmHWHM*2)
```

```python
for i in range(px):
    for j in range(py):
        nvintensity[i][j] = getNvIntensity(data, i, j)
```

```python
plt.figure(figsize=(12,9))
plt.grid()
plt.imshow(nvintensity, cmap='viridis')
plt.colorbar()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

# Bin SiV-

```python
def getSivIntensity(data, regionx, regiony):
    currentPL = data['im'][regionx][regiony]
    model2 = QuadraticModel(prefix='bkg_') + LorentzianModel(prefix='sivm_')
    model = model2
    params = model.make_params()

    params['sivm_center'].set(value=739, min = 738, max = 740, vary=True)
    params['sivm_amplitude'].set(value=1000, min = 0, vary=True)
    params['sivm_sigma'].set(value=2, max = 5, vary=True)

    params['bkg_a'].set(value = 0, vary=True)
    params['bkg_b'].set(value = 0, min = 0, vary=True)
    params['bkg_c'].set(value = 0, vary=True)

    Index = list(range(1900,2400))
    weightList = 1/np.sqrt(currentPL[Index])
    result = model.fit(currentPL[Index], params, x=wavel[Index], weights=weightList)
    comps = result.eval_components(x=wavel[Index])

    sivmCenter = result.params['sivm_center']
    sivmHWHM = result.params['sivm_sigma']
    print(regionx,regiony)
    return integrate(wavel[Index],comps['sivm_'],sivmCenter - sivmHWHM*2, sivmCenter + sivmHWHM*2)
```

```python
for i in range(px):
    for j in range(py):
        sivintensity[i][j] = getSivIntensity(data, i, j)
```

```python
plt.figure(figsize=(12,9))
plt.grid()
plt.imshow(sivintensity, cmap='viridis')
plt.colorbar()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

# Bin second order Raman

```python
def fcn2min(params, x, data):
    """Model a decaying sine wave and subtract data."""
    scale = params['scale']
    a = params['a']
    b = params['b']
    c = params['c']
    model = a*x**2+b*x+c
    return model + scale * ramanI[566:836] - data
```

```python
def getR2Intensity(data, regionx, regiony):
    Index = list(range(560,830))
    currentPL = data['im'][regionx][regiony]
    # create a set of Parameters
    params = Parameters()
    params.add('scale', value=1, min=0)
    params.add('a', value=0)
    params.add('b', value=0.0)
    params.add('c', value=0)

    # do fit, here with the default leastsq algorithm
    minner = Minimizer(fcn2min, params, fcn_args=(wavel[Index], currentPL[Index]))
    result = minner.minimize()
    print(regionx, regiony)
    return result.params['scale'].value
```

```python
for i in range(px):
    for j in range(py):
        r2intensity[i][j] = getR2Intensity(data, i, j)
```

```python
plt.figure(figsize=(12,9))
plt.grid()
plt.imshow(r2intensity, cmap='viridis')
plt.colorbar()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
