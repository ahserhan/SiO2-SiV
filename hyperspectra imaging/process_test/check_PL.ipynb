```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm

def findarg(data, value):
    arg = 0
    for wl in data:
        if wl<value:
            arg = arg + 1
        else:
            return arg
        
    return arg




regionMin = 600
regionMax = 622.1

```

```python
data = pd.read_csv('sin_m3-10_offSquare_Full_2mW_30s 2023 June 19 16_42_38.csv')
argMin = findarg(data['W'], regionMin)
argMax = findarg(data['W'], regionMax)

plt.figure(figsize=(12,9))
plt.grid()
plt.plot(data['W'][argMin:argMax],data['I'][argMin:argMax])
```

```python
data = pd.read_csv('sio_offSquare_Xis0_Full_2mW_30s 2023 June 20 14_13_29.csv')
argMin = findarg(data['W'], regionMin)
argMax = findarg(data['W'], regionMax)

plt.figure(figsize=(12,9))
plt.grid()
plt.plot(data['W'][argMin:argMax],data['I'][argMin:argMax])
```

```python
norm = data['I'] - data['I'][argMax-1]
plt.figure(figsize=(12,9))
plt.grid()
plt.plot(data['W'][argMin:argMax],norm[argMin:argMax])
```

```python
r2 = []
for i in range(0,argMin):
    r2.append(0)
for i in range(argMin,argMax):
    r2.append(norm[i])
for i in range(argMax,len(norm)):
    r2.append(0)

plt.figure(figsize=(12,9))
plt.grid()
plt.plot(data['W'],r2)
```

```python
df = pd.DataFrame({'I': r2, 'W':data['W'] })
df.to_csv('secondRaman.csv')
```
