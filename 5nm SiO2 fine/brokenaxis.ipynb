```python
import numpy as np
import matplotlib.pyplot as plt
from brokenaxes import brokenaxes

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = 5 * np.cos(x) + 10  # Adjusted for a different range

# Create a brokenaxes instance with subplot_spec parameter
fig = plt.figure()
# ax = fig.add_subplot(111)

# Set the position of the brokenaxes instance in the figure
bax = brokenaxes(ylims=((0, 5), (10, 15)), hspace=.05, despine=False)

# Plot the two ranges of data
bax.plot(x, y1, label='Data 1', color='blue')
bax.plot(x, y2, label='Data 2', color='orange')

# Add labels and legend
bax.set_xlabel('X-axis')
bax.set_ylabel('Y-axis')
# bax.legend(loc='upper left')

# Show the plot
plt.show()
```
