```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm

def findarg(data, value):
    arg = 0
    for wl in data:
        if wl<value:
            arg = arg + 1
        else:
            return arg
        
    return arg




regionMin = 550
regionMax = 575

```

```python
data = pd.read_csv('sio_m3-9_Xis0_Full_2mW_30s 2023 June 20 11_29_23.csv')
argMin = findarg(data['W'], regionMin)
argMax = findarg(data['W'], regionMax)

plt.figure(figsize=(12,9))
plt.grid()
plt.plot(data['W'][argMin:argMax],data['I'][argMin:argMax])
```
