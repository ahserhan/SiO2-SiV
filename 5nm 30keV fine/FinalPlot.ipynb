```python
import numpy as np 
import matplotlib.pyplot as plt 
import uncertainties as unc
import pandas as pd
import uncertainties.unumpy as unp
```

```python
SiO = pd.read_csv('SiO.csv')
SiN = pd.read_csv('SiN.csv')
SiOfine = pd.read_csv('SiOfine.csv')
SiNfine = pd.read_csv('SiNfine.csv')
```

```python
plt.figure(figsize=(9,6.5))
plt.grid()
plt.xlabel(r'Depth in Diamond (nm)')
plt.ylabel(r'Density per Dose (cm$^{-3}$/cm$^{-2}$)')
plt.plot(SiO['DEPTH'], SiO['Ga'], '--k',label = 'Ga')
plt.plot(SiO['DEPTH'], SiO['C']*1e8, 'ob',label = 'Vacancy')
plt.plot(SiOfine['DEPTH'], SiOfine['Si'], '-r',label = 'Si')
plt.plot(SiOfine['DEPTH'], SiOfine['O'], '-g',label = 'O')
plt.title('5 nm SiO$_2$ Layer with 30 keV Ga$^{+}$ FIB')
plt.legend()

a = plt.axes([0.45, .3, .4, .4])
# plt.grid()
plt.plot(SiO['DEPTH'], SiO['Ga'], '--k')
plt.plot(SiOfine['DEPTH'], SiOfine['Si'], '-r')
plt.plot(SiOfine['DEPTH'], SiOfine['O'], '-g')
plt.xlim([0,25])
plt.ylim([0,0.6e7])
# plt.title('Probability')
# plt.xticks([])
# plt.yticks([])
```
