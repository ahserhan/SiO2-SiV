```python
import numpy as np 
import matplotlib.pyplot as plt 
import uncertainties as unc
import pandas as pd
import uncertainties.unumpy as unp
from matplotlib.cbook import get_sample_data
```

```python
SiO = pd.read_csv('SiO.csv')
SiN = pd.read_csv('SiN.csv')
SiOfine = pd.read_csv('SiOfine.csv')
SiNfine = pd.read_csv('SiNfine.csv')
```

```python
cm = 1/2.54
fig = plt.figure(figsize=(20*cm, 15*cm))
plt.rcParams.update({'font.size': 22})

# plt.rcParams.update({'font.size': 6})
# cm = 1/2.54 
# fig = plt.figure(figsize=(4.5*cm, 3.4*cm))
# plt.grid()
plt.xlabel(r'Depth in Diamond (nm)')
plt.ylabel(r'Density per Dose (cm$^{-3}$/cm$^{-2}$)')
plt.plot(SiO['DEPTH'], SiO['Ga'], '--k',label = 'Ga')
plt.plot(SiO['DEPTH'], SiO['C']*1e8, 'oC0',label = 'Vacancy')
plt.plot(SiOfine['DEPTH'], SiOfine['Si'], '-C2',label = 'Si')
plt.plot(SiOfine['DEPTH'], SiOfine['O'], '-C3',label = 'O')
plt.title('5 nm SiO$_2$')
# plt.legend()

a = plt.axes([0.57, .33, .37, .45])
# plt.grid()
plt.plot(SiO['DEPTH'], SiO['Ga'], '--k')
plt.plot(SiOfine['DEPTH'], SiOfine['Si'], '-C2')
plt.plot(SiOfine['DEPTH'], SiOfine['O'], '-C3')
plt.xlim([0,25])
plt.ylim([0,0.6e7])
# plt.title('Probability')
# plt.xticks([])
# plt.yticks([])

plt.plot([], '--k',label = 'Ga$^+$')
plt.plot([], 'oC0',label = 'Vacancy')
plt.plot([], '-C2',label = 'Si')
plt.plot([], '-C3',label = 'O')

lgd = plt.legend(loc = "upper right")
lgd.get_frame().set_alpha(0.1)
plt.tight_layout()
plt.savefig('sio.eps', format= 'eps',transparent=False)
```
