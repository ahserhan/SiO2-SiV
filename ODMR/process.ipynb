```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm
```

```python
outSquare = pd.read_csv('out.csv')
inSquare = pd.read_csv('in.csv')
```

```python
plt.plot(outSquare['xdata']/1e9,outSquare['ydata'], '-')
plt.xlim([2.7,3.04])
plt.yticks([0.99,0.9925,0.995,0.9975, 1], [0.99,0.9925,0.995,0.9975, 1])
```

```python
plt.plot(inSquare['xdata']/1e9,inSquare['ydata'])
plt.xlim([2.7,3.04])
plt.yticks([0.9925,0.995,0.9975, 1],[0.9925,0.995,0.9975, 1])
```

```python
plt.rcParams.update({'font.size': 6})
cm = 1/2.54 
fig, ax = plt.subplots(figsize=(4*cm, 3.5*cm))
# fig = plt.figure(figsize=(2*cm, 1.5*cm))
# plt.grid()
# plt.title('SiV Spectra')
plt.xlabel(r'RF Frequency (GHz)')
plt.ylabel(r'PL Contrast')
# plot the data with errorbars, and to see the errorbars, we multiply it by 100
plt.plot(inSquare['xdata']/1e9,inSquare['ydata'], '-C1', linewidth = 0.5)
plt.xlim([2.7,3.04])
plt.yticks([0.992,0.996, 1], [0.992,0.996, 1])
# plt.legend()
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(0.1)

plt.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,
    left = False,
    right = False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off
ax.tick_params(axis="y",direction="in")
ax.tick_params(axis="x",direction="in")

plt.tight_layout()

fig.savefig('inSquare.eps', format= 'eps',transparent=True)
```
