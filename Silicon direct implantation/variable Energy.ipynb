```python
def plot3dSilicon(DEPTH,data,thickness,lab):
    if lab == 0:
        ax.plot(DEPTH, data, zs=thickness, zdir='y', lw = 1, color = 'r')
    else:
        ax.plot(DEPTH, data, zs=thickness, zdir='y', lw = 1, color = 'r', label = 'Silicon Atom Distribution')
```

```python
def plot3dVacancy(DEPTH,data,thickness,lab):
    if lab == 0:
        ax.plot(DEPTH, data, zs=thickness, zdir='y',  color = 'b' )
    else:
        ax.plot(DEPTH, data, zs=thickness, zdir='y',  color = 'b',  label = 'Carbon Vacancy Distribution')
```
