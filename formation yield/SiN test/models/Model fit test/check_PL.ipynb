```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm

def findarg(data, value):
    arg = 0
    for wl in data:
        if wl<value:
            arg = arg + 1
        else:
            return arg
        
    return arg


filename = 'sin_m3-15_Xis0_Full_0p2mW_100s 2023 June 19 17_34_07.csv'

regionMin = 783.7
regionMax = 784.5

```

```python
data = pd.read_csv(filename)
argMin = findarg(data['W'], regionMin)
argMax = findarg(data['W'], regionMax)

plt.figure(figsize=(12,9))
plt.grid()
plt.plot(data['I'])
```
