```python
def calculate_variance(x_values, y_values):
    mean_x = np.average(x_values, weights=y_values)
    mean_x2 = np.average(x_values**2, weights=y_values)
    variance = mean_x2 - mean_x**2
    return mean_x,np.sqrt(variance)
```
